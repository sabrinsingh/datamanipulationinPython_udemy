Creating DataFrame using a list

```python
#one dimensional list
lst= ['apple','banana']
```

```python
#two dimensional list
lst= [ ['apple','mango'], ['100','60']   ]

df=pd.DataFrame(lst, columns=['fruits','price'])
df
```

Creating dataframe using dictionary

```python
myDict = {
    'fruits' :['apple','mango'],
    'price' :[100,60]
}
df=pd.DataFrame(myDict)
df
```
