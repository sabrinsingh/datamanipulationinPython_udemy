```python
import numpy as np
import pandas as pd
```

Median based Anomaly Detection

```python
x = pd.Series([2.1,2.3,4.5,2.2,2.4])

median = np.median(x)
threshold = 2
outliers = []
for item in x:
    if abs(item - median)>threshold:
        outliers.append(item)

print(outliers)
```

Mean based Anomaly Detection

```python
x = pd.Series([2.1,2.3,4.5,2.2,2.4])

avg = np.mean(x)
std = np.std(x)
print(f'Average is {avg} and Standard Deviation is {std}')

outliers = []
for item in x:
    if (item < avg -std) or (item >avg + std):
        outliers.append(item)

print(outliers)
```

Z-score based anomaly detection

```python
x= pd.Series([2.1,2.3,4.5,2.2,2.4])
print(x)
```

```python
mean     = np.mean(x)
std      =  np.std(x)
outliers = []
for item in x:
    z_score = (item - mean) /std
    if z_score > 1.5:
        outliers.append(item)
print(outliers)        
    
```

Regular Expressions or RegEx----for this we use re module

```python
import re
```

```python
text = 'Python is my favorite programming language. I love Python'
x=re.findall('Python', text)
print(x)
```

To check if the string starts with certain pattern we can use hat ^

```python
y=re.findall('^Python',text)
print(y)
```

```python
text1 = 'It is Python'
y=re.findall('^Python',text1)
print(y)
```

To findall all the digit in string --->findall('\d',str)

```python
import re

# Sample text
text = "There are 3 apples and 5 oranges."

# Find all digits in the text
digits = re.findall('\d', text)

print(digits)
```
